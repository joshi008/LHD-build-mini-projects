```python
import requests
import matplotlib.pyplot as plt

x = requests.get('https://codeforces.com/api/problemset.problems')
data = x.json() 
data = data['result']['problems']

tags = set()
count={}
for problem in data:
    for tag in problem['tags']:
        if(tag not in count):
            count[tag]=1
        else:
            count[tag]+=1
# print(count)


plt.bar(range(len(count)), list(count.values()), align='center')
plt.xticks(range(len(count)), list(count.keys()), rotation='vertical')

plt.show()
```
